1) What is hypothesis testing in statistics ?

Ans:- hypothesis testing is a crucial method for making informed decisions about a population based on sample data.

2) What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans:- The null hypothesis assumes no effect or relationship between variables, while the alternative hypothesis proposes a relationship or effect exists. The null hypothesis is what researchers aim to disprove, while the alternative hypothesis is what they hope to find evidence for, meaning the alternative hypothesis is what they are trying to prove.



3) What is the significance level in hypothesis testing, and why is it important ?

Ans:-  the significance level (α) is the probability of rejecting the null hypothesis when it is actually true. It essentially defines the threshold of evidence required to reject the null hypothesis. Why is it important? It controls the risk of making a Type I error (false positive), ensuring the validity and reliability of the findings.


4) What does a P-value represent in hypothesis testing ?

Ans:-  p-value represents the probability of observing data as extreme as or more extreme than the observed results, assuming that the null hypothesis is true. It's essentially a measure of the strength of evidence against the null hypothesis.


6) What are Type 1 and Type 2 errors in hypothesis testing ?

Ans:- a Type I error occurs when a true null hypothesis is rejected, while a Type II error occurs when a false null hypothesis is not rejected


7) What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?

Ans:- a one-tailed test checks if a parameter is greater than or less than a specific value, while a two-tailed test checks if it's different from that value (either greater or less). One-tailed tests focus on one direction, while two-tailed tests consider both directions of deviation.


8) What is the Z-test, and when is it used in hypothesis testing ?

Ans :- A z-test is a statistical test used to determine if there's a significant difference between a sample mean and a population mean, or between the means of two groups when the population variance is known and the sample size is large (typically n > 30)

9) How do you calculate the Z-score, and what does it represent in hypothesis testing ?

Ans:- To calculate a z-score, subtract the population mean from a data point and divide the result by the population standard deviation. In hypothesis testing, the z-score represents how many standard deviations a data point or sample mean is from the hypothesized population mean. It's used to determine the probability of observing a sample mean as extreme as or more extreme than the one observed, assuming the null hypothesis is true.

10) What is the T-distribution, and when should it be used instead of the normal distribution ?

Ans:- The t-distribution is a probability distribution similar to the normal distribution but with heavier tails, meaning it has a greater chance of extreme values. It's used as an alternative to the normal distribution when dealing with small sample sizes (n<30) and when the population standard deviation (σ) is not known.

11) What is the difference between a Z-test and a T-test ?

Ans:- The main difference between a z-test and a t-test lies in their use of population standard deviation (σ) and sample size. Z-tests are used when the population standard deviation is known or when the sample size is large (generally > 30). T-tests are used when the population standard deviation is unknown, especially when the sample size is small (generally < 30).


12) What is the T-test, and how is it used in hypothesis testing ?

Ans:- A t-test is an inferential statistic used to determine if there is a statistically significant difference between the means of two population samples.It is used in statistics for hypothesis testing and can indicate whether differences between two populations are meaningful or random.



13) What is the relationship between Z-test and T-test in hypothesis testing ?

Ans:- The Z-test and T-test are both statistical hypothesis tests used to compare means, but they differ in their assumptions about sample size and population variance. The Z-test is typically used for large samples (n > 30) when the population standard deviation is known. The T-test is used for small samples (n < 30) or when the population standard deviation is unknown.



14) What is a confidence interval, and how is it used to interpret statistical results ?

Ans:-A confidence interval is a range of values used to estimate an unknown population parameter based on sample data. It provides a level of confidence (e.g., 95%, 99%) that the true population parameter falls within that range. This range reflects the uncertainty in the estimate due to sampling variability.



15) What is the margin of error, and how does it affect the confidence interval ?

Ans:- The margin of error (ME) indicates how much a sample statistic (like a survey result) might differ from the true population parameter. It's essentially half the width of a confidence interval. A smaller ME means the estimate is more precise, while a larger ME suggests greater uncertainty. The ME directly impacts the confidence interval, affecting its width and therefore the level of confidence in the estimate.



16) How is Bayes' Theorem used in statistics, and what is its significance ?

Ans:- Bayes' Theorem is a fundamental concept in probability and statistics, used to update the probability of a hypothesis (event) given new evidence. It allows for the revision of initial beliefs (prior probabilities) based on observed data, resulting in posterior probabilities.


17) What is the Chi-square distribution, and when is it used ?

Ans:- The chi-square distribution is a continuous probability distribution that is often used in hypothesis testing, specifically for categorical data. It's used to determine if there's a statistically significant difference between observed frequencies and expected frequencies, or to test the independence of two categorical variables.



18) What is the Chi-square goodness of fit test, and how is it applied ?

Ans:- The chi-square goodness of fit test is a statistical test used to determine how well a sample distribution matches a hypothesized distribution. It assesses whether the observed frequencies of categories in a sample data deviate significantly from the expected frequencies based on the hypothesized distribution.



19) What is the F-distribution, and when is it used in hypothesis testing ?

Ans:- The F-distribution is a continuous probability distribution used in hypothesis testing, particularly when comparing variances between two or more populations


20) What is an ANOVA test, and what are its assumptions ?

Ans:- ANOVA, or Analysis of Variance, is a statistical test used to compare the means of two or more groups to determine if there's a significant difference between them
1. Independence of Observations: Data points within each group must be independent of each other
2. Normal Distribution: The population distribution from which the samples are drawn should be normally distributed.
3. Homogeneity of Variance: The variances (spread) of the populations from which the samples are drawn should be equal. This means the variability within each group should be similar.
4. Random Sampling: The observations within each group should have been randomly sampled. This helps ensure that the groups are representative of their respective populations.



21) What are the different types of ANOVA tests ?

Ans:- The two main types of ANOVA tests are one-way ANOVA and two-way ANOVA. One-way ANOVA compares the means of two or more groups when there is only one independent variable. Two-way ANOVA, also known as factorial ANOVA, is used when there are two or more independent variables, and the researchers are interested in examining the main effects of each independent variable and also their interaction effects.

22) What is the F-test, and how does it relate to hypothesis testing?

Ans:- The F-test is a statistical test that compares variances between two or more groups or samples, often used as part of Analysis of Variance (ANOVA). It helps determine if there's a significant difference in the means of these groups, by assessing the ratio of variances between the groups and the variance within the groups. In hypothesis testing, the F-test helps decide whether to reject the null hypothesis, which often states that there's no difference between the population variances or means.

```python
# 15) Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    reject = p_value < alpha
    return z_stat, p_value, reject

sample_data = np.array([102, 99, 101, 98, 100, 97, 103, 104, 99, 101])
pop_mean = 100
pop_std = 5
alpha = 0.05

z_stat, p_value, reject = z_test(sample_data, pop_mean, pop_std, alpha)
print(f'Z-statistic: {z_stat:.3f}')
print(f'P-value: {p_value:.4f}')
print('Reject null hypothesis' if reject else 'Fail to reject null hypothesis')
```

```python
# 16) Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = sem * stats.t.ppf((1 + confidence) / 2., n - 1)
    lower = mean - margin
    upper = mean + margin
    return lower, upper

data = np.array([102, 99, 101, 98, 100, 97, 103, 104, 99, 101])
ci_lower, ci_upper = confidence_interval(data)

print(f'{int(100*0.95)}% confidence interval: ({ci_lower:.2f}, {ci_upper:.2f})')

```

```python
# 1) Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    reject = p_value < alpha
    return z_score, p_value, reject

sample = np.array([105, 102, 98, 101, 99, 100, 104, 106, 97, 103])
population_mean = 100
population_std = 5

z_score, p_value, reject = one_sample_z_test(sample, population_mean, population_std)

print(f'Z-score: {z_score:.3f}')
print(f'P-value: {p_value:.4f}')
print('Reject null hypothesis' if reject else 'Fail to reject null hypothesis')
```

```python
# 3) Implement a one-sample Z-test using Python to compare the sample mean with the population mean
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    reject = p_value < alpha
    return z_stat, p_value, reject

sample = np.random.normal(loc=102, scale=5, size=30)
population_mean = 100
population_std = 5

z_stat, p_value, reject = one_sample_z_test(sample, population_mean, population_std)
print(f'Z-statistic: {z_stat:.3f}')
print(f'P-value: {p_value:.4f}')
print('Reject null hypothesis' if reject else 'Fail to reject null hypothesis')
```

```python
# 4) Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_two_tailed_z_test(sample, population_mean, population_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    z_critical = norm.ppf(1 - alpha / 2)

    plt.plot(x, y)
    plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3)
    plt.axvline(z, color='black', linestyle='--', label=f'Z = {z:.2f}')
    plt.title('Two-Tailed Z-Test Decision Region')
    plt.xlabel('Z')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z, 2 * (1 - norm.cdf(abs(z)))

sample = np.random.normal(102, 5, 30)
z_stat, p_val = plot_two_tailed_z_test(sample, 100, 5)
print(f'Z-stat: {z_stat:.2f}, P-value: {p_val:.4f}')
```

```python
# 5) Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
def visualize_type1_type2(mu0, mu1, sigma, n, alpha):
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    se = sigma / np.sqrt(n)
    z_critical = norm.ppf(1 - alpha)
    threshold = mu0 + z_critical * se

    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.plot(x, y0, label='H0 Distribution')
    plt.plot(x, y1, label='H1 Distribution')
    plt.fill_between(x, y0, where=(x >= threshold), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, y1, where=(x < threshold), color='blue', alpha=0.3, label='Type II Error')
    plt.axvline(threshold, color='black', linestyle='--', label='Critical Value')
    plt.title('Type I and Type II Errors')
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_type1_type2(100, 105, 10, 30, 0.05)

```

```python
# 9) Write a Python function to calculate the confidence interval for a sample mean and explain its significance
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = sem * t.ppf((1 + confidence) / 2., n - 1)
    return mean - margin, mean + margin

data = np.random.normal(50, 5, 30)
ci = confidence_interval(data)
print(f'95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})')
```

```python
# 10) Write a Python program to calculate the margin of error for a given confidence level using sample data
def margin_of_error(data, confidence=0.95):
    n = len(data)
    sem = stats.sem(data)
    moe = sem * t.ppf((1 + confidence) / 2., n - 1)
    return moe

data = np.random.normal(70, 10, 40)
moe = margin_of_error(data)
print(f'Margin of Error: {moe:.2f}')
```

```python
# 11) Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_A, likelihood_B_given_A, prob_B):
    posterior_A_given_B = (likelihood_B_given_A * prior_A) / prob_B
    return posterior_A_given_B

prior = 0.01
likelihood = 0.9
false_positive = 0.05
prob_B = prior * likelihood + (1 - prior) * false_positive

posterior = bayes_theorem(prior, likelihood, prob_B)
print(f'Posterior Probability: {posterior:.4f}')
```
